```python
import numpy as np
import porespy as ps
import scipy
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
np.random.seed(10)
```

perlin_noise creates an image of a perlin noise field. 

```python
shape = [1000,1000]
porosity = None
octaves = 2
frequency = [10,10]
persistence = 0.5
```

```python
x = ps.generators.perlin_noise(shape, porosity, octaves, frequency, persistence)
print(x)
```

```python
fig = plt.imshow(x);
```

While adding inputs to the function, the frequency must be a factor of shape, that when put to the power of octave is less than or equal to shape. If not, exceptions will be raised:

```python
shape = [1000,1000]
porosity = None
octaves = 4
frequency = [10,10]
persistence = 0.5
```
